Output formating:

Types:
    1.old style formatting(using comma)
    2.access specifiers formating
    3.dot format
    4.f-format


```python
user = input()
age = int(input())
print("my name is", user,"and my age is",age) 
```

```python
#f format
name = "balaji"
age = 23
print(f"my name is {name} and i am {age} years old")
```

```python
#dot format
name = "balaji"
age = 23
print("my name is {} and i am {}years old".format(name,age))
```
