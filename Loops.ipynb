LOOP:
    -perform same operation repeatedly again and again is called loop.

TYPES OF LOOPS:
    -1.for loop
    -2.while loop

1.for loop:                               #for - keyword
    syntax:                               #iterator - variable
    >for iterator in iterable:            #in - membership op
        #block code                       #iterable - list,sets,tuple,strings,dic,range()
     else:(optional)
         #block code

>for loop - we use when you know total iterations and this loop can use iterate over the list,tuple,set,dic,string and range() function.
>automatic increment

```python
#check wheather it is even or odd
lst = [12,13,14]
for num in lst:
    if (num%2==0):
        print("even")
    else:
        print("odd")
```

```python
s = "codegnan" 
for each_char in s:
    print(each_char)
```

```python
list1 = list(map(int, input().split()))
for num in list1:
    if (num%2==0):
        square = (num*num)
        print(square)
    else:
        num = num + 1
        print(num)
```

```python
#to check the count of vowels and consonants in the given string
string = input().lower()
c = v = 0
for char in string:
    if char.isalpha():   
        if char in 'aeiou':
            v += 1
        else:
            c += 1            
print(v,c)
```

range() :

syntax: 
    range(start,end,step)


```python
n = int(input())
for i in range(1,n+1):
    print(i)
```

```python
n = int(input())
for i in range(1,n+1):
    print(i*i*i)
```

```python
n = int(input())
for i in range(1,n+1):
    print("*",end='\n')
```

```python
n = int(input())
for i in range(1,n+1):
    print("*",end=" ")
```

```python
n = int(input())
for i in range(1,n+1):
    print("*",end="hi")
```

WHILE LOOP: The loop body executes until condition evaluates false.
>in while loop manual increment or decrement
>if you dont know the total no of iterations we will use while loop

Syntax:

#initilization
while condition:
    #block code
    #increment
else:
    #block code


```python
n = int(input())
i = 1 #initilization
while i<=n:
    print(i)
    i+=1
```

```python
n = int(input())
i = 1 #initilization
while i<=n:
    print(n)
    n-=1
```
