```python
#Define list
#it is ordered collection
list1 = [1,2,3,4,'hi',1,3,4]
print(list1)
```

```python
#indexing operations
list1 = [1,2,4.5,'hello',34]
print(list1[2])
list1[2] = list1[3]
print(list1[2])
print(list1[-1])
```

```python
lst1 = [10,20,30,40,50,60,70]
print(lst1[:5:-1])
print(lst1[:500])
#reverse list
print(lst1[::-1])
```

```python
#Empty listcreation
l1 = []
l2 = list()
print(type(l1),type(l2))
```

```python
#List Operations
#1. List concatination
#2. List repetition (using * operator)

l1 = [1,2,3]
l2 = [3,4,5,6,7]
print(l1+l2)
print(l1 * 3)
print([0] * 10)
```

```python
# List methods
-Item addding methods
    -append(item): It adds item at end of list
    -extend(iterable): It adds the all iterator elements at end of the list
    -insert(pos,item): it adds item at specified position
```

```python
lst1 = [10,20,30,40]
lst1.append(500)
print(lst1)
lst1.extend([1,2,3])
print(lst1)
lst1.append([100,200])
print(lst1)
lst1.insert(0,1000)
##it adds at 0 index position and then remaining elements are move right shift 
print(lst1)
```

```python
#Item Removing Methods
 -remove(item): It removes the first occurance item from the list if it is exists in list otherwise 'Value error'
 -pop(pos) : It removes and return the specified position item if position is passed otherwise it remove and returns the last item from the list.
 -clear(): It removes all elements from list
```

```python
lst1 = [10,20,30,40]
lst1.remove(20)
print(lst1)
```

```python
lst = [10,20,30,20,40,50,20]
val = lst.pop() # it removes the last element of the list
print(val)
print(lst)
val2 = lst.pop(0) #it removes the 0th position
print(val2)
print(lst)
lst.clear() #it cleares the list
print(lst)
```

```python
#index and count()
#index(item): if item is present in list it returns first occurance the index, otherwise ValueError
#counr(item): It returns the count of total occurances of item
lst = [10,20,30,20,40,50,20]
ind = lst.index(20)
print(ind)
print(lst.count(20))
print(lst.count(60)) #it returns 0 if value is not in the list.
```

```python
#Sorting items
#sort(reverse = False): By default it sorts the original list in into ascending order. 
#                         If reverser is True it sorts in decending order.
lst = [1,2,3,4,4.5,-45,-4,56,78,90,103]
lst.sort()     #ascending order
print(lst)
lst.sort(reverse = True)   #decending order
print(lst)
```

```python
lst = ['hey','balaji','lasya',]
lst.sort()     #ascending order
print(lst)
```

```python
#Built in Functions: len(),min(),max(),sum()

lst = [1,2,3,5,6]
print(len(lst))
print(max(lst))
print(min(lst))
print(sum(lst))
print(sum(lst)/len(lst))   #to find the avg(sum of elements/length of list)
```

```python
lst = ['hey','balaji','lasya',]
print(len(lst))
print(max(lst))
print(min(lst))
```

## TUPLE:

    defines with '( )'
Characteristics:
    -is ordered collection
    -is immutable
    -allows indexing and slicing
    -is heterogeneos datatype
    -allows duplicate value

##LISTS:                        ##TUPLES
    -mutable                        -immutable
    -represents [ ]                 -represens ( )
    -lists takes more memory space  -less memory space
    -operation speed less           -operation speed more

## Nested Indexing:

```python
t = [1,2,[4,5,6,[7,9,10],11],12]
print(t[2])
```

```python
t = [1,2,[4,5,6,[7,9,10],11],12]
print(t[3])
print(t)  ##ordered collection
```

```python
##Immutable
t = (1,2,3,[10,20,30,40],23)
t[3].append(200)
print(t) ##it does not change the obj length and does not change the remaining element positions
```

```python
#EMPTY TUPLE REPRESENTATION
t1 = ()
t2 = tuple()
print(type(t1),type(t2))
```

```python
#SINGLE ELEMENT REPRESENTATION
t1 = (10,) #we use ',' then only it shows type as 'tuple' if not mentioned it shows 'int'
l1 = [10]
print(type(t1),type(l1))
```
