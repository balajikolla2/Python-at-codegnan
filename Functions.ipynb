Types of arguments:

```python

#1.Positional arguments: Here arguments are assigns to parameters as per the sequential 

#subraction

#function defination block
def sub(x:int,y:int)->int:
    z = x-y
    return z

#function calling block
a,b = 20,10
print(sub(a,b))
print(sub(b,a))
```

```python
#2. Keyword arguments: arguments are explicitly passed with parameter names

#function defination block
def sub(x:int,y:int)->int:
    z = x-y
    return z

#function calling block
a,b = 20,10
print(sub(x = a,y = b))
print(sub(y = b,x = a))
```

```python
def add(x:int,y:int,z:int)->int:
    z1 = x+y+z
    return z1

#function calling block
a,b,c= 20,10,20
print(add(x = a,y = b,z = c))
```

```python
def add(x:int,y:int,z:int,r:int)->int:
    o = x+y+z+r
    return o

#function calling block
a,b,c,d= 20,10,20,10
print(add(x = a,y = b,z = c,r = d))
```

```python
def add(x:int,y:int,z:int,r:int)->int:
    res = x+y+z+r
    return res

a,b,c,d = map(int,input().split())
print(add(a,b,c,d))
```

```python
#for different operations same function

def add(w:int,x: int,y:int =0,z: int =0)->int:
    res = w + x + y + z 
    return res                            

a,b,c,d = map(int,input().split())            #here if any updated value is exist then it will execute or then default value will exist, for y and z.
print(add(a,b))           
print(add(a,b,c))
print(add(a,b,c,d))
```

```python
#variable length positional arguments 

def add(x,y =0,*args):                 #for many arguments like 2 ,3 ,..... we use *args , no need to define the parameters for all the arguments.
    print(type(args))
    print(args)                        # * single: for positional args and ** for keyword args
    s = 0 
    for val in args:
        s += val
    return s

print(add(10,20,1,2,3,4,5))
```

```python
#variable length keyword arguments

def add(x,y=0, * args, **kwargs):    
    print(type(kwargs))
    print(args)
    print(kwargs)
    print(kwargs.values())
    print(kwargs.keys())

print(add(1,2,3,4,5,name = "balaji", age = 20, course = "PFS"))
```

```python
#In python every type function calling block shares the arguments reference to parameter
#If the arguments are immutable data types,then behave likes pass by value and if
#arguments are mutable datatypes(i.e, list,set,dict) then behave likes pass by reference  

#pass by value

def fun(x):
    print(id(x))
    x += 2
    print("after add 2")
    print(id(x))
    print(x)

a = 10
print("Before function execution 'a' value:", a)
fun(a)
print("After function execution 'a' value:", a)
print(a)
```
