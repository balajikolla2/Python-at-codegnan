Conditional Statements: to make decision
Types: 
    1. if
    2.if-else
    3.if-elif-else
    4.nested if

1.if statement: The block code is executes when if condition evaluates to true.
    syntax:
        if condition:
            block code

```python
a = int(input("Check your eligibility:" ))
if (a>=18):
              print("Congratulations you are eligible!")
print("Next year")   
```

```python
## 2.if else: The condition is true it executes if block if it is false then else block executes.
syntax:
if condition:
    block of code
else:
    block of code
```

```python
a = int(input())
if(a>=18):
    print("you are eligible")
else:
    print("next year buddy!")
```

```python
bill_amount = int(input())
if bill_amount > 500:
    discount_price = bill_amount * 10/100
    final_bill = (bill_amount-discount_price)
    print("Hey your are final amount after discount:",final_bill)
else:
    print("No discount your bill is below 500:",bill_amount)
```

```python
 #5
products = {'pen': 5, 'book': 20, 'pencil':3}
item = input().lower() #it converts string into lower case for to overcome errors
quantity = int(input())
#by using membership to check it is in that existing values or not
if item in products:
    total_amount = products.get(item) * quantity
    print(total_amount)
else:
     print("item out of stock")
```

if else elif: The if block code executes when if condition evaluates to true,otherwise it sequentially checks with elif condtions,if any condition is true,then it executes respective block,otherwise else will execute

syntax:
if condition1:
    #block1
elif condition2:
    #block2
elif condition3:
    #block3
it can have multiple elif statements (but only one else block and no condition in else block is optional)

```python
#check weather the num is pos,neg,or zero
n = int(input())
if(n>0):
    print("positive")
elif(n<0):
    print("negative")
else:
    print("zero")
```

```python
n = int(input())
if(n%3==0):
    print("Fizz")
elif(n%5==0):
    print("Buzz")
else:
    print("Fizzbuzz")
```
