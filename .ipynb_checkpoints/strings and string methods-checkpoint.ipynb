```python
##string methods
#- case conversion methods
#    -upper(): it converts entire string into uppercase
#    -lower(): it converts entire string into lowercase
#    -title(): it converts first letter of every word into upper and remain lower
#    -swapcase(): it coverts upper to lower or lower to upper case
#    -capitalize(): it converts first char to upper and remain lower
```

```python
s1 = "hey how are you"
print(s1.upper())
print(s1.lower())
print(s1.title())
print(s1.swapcase())
print(s1.capitalize())

print(s1) # immutable nature
```

```python
#Remove white space or pattern at ends of string
    -strip(): By default it removes the white space at both ends of string
    -lstrip(): removes left space white spaces
    -rstrip(): removes right space white spaces
```

```python
s1 = "codegnan"
print(s1)
print(s1.strip())
print(s1.lstrip())
print(s1.rstrip())
```

```python
s1 = "$$$$$$codegnan$$$$$$"
print(s1)
print(s1.strip("$"))
print(s1.lstrip("$"))
print(s1.rstrip("$"))
```

```python
# split(): it splits the entire string into list of sub strings based on the specified
#by default it takes spaces
```

```python
s1 = "The sky is blue"
print(s1.split("is"))
print(s1.split("e"))
```

```python
words = ['The' , 'sky', 'is' , 'blue']
print("".join(words))
print(" ".join(words))
print("_".join(words))
```

```python
#Substring index finding
#in index: if char is not found it returns value error
#in find : it returns "-1" for no char is exists in the string
```

```python
s1 = "The sky is blue"
print(s1.index('is'))
print(s1.find('are'))
```

```python
s1 = "The sky is blue and ocean is blue"
print(s1.find('is'))
print(s1.find('are'))
```

```python
#count(substring): it returns the total non-overlapping occurences of substring
s = "ABCABBabbbab"
print(s.count('B'))
print(s.count('BCB'))
print(s.count('Z'))
```

```python
#replace(old_substring,new_substring): It replace the old sub string with new sub string
s1 = "The sky is blue and blue and ocean is blue"
print(s1.replace("blue","red"))
print(s1.replace('e','E',2)) # it replace the first 2 occurances
```

```python
#string checking methods it gives boolean values
-isupper(): it returns true, if entire string is in uppercase
-islower(): it returns true, if entire string is in lowercase
-isalpha(): it returns true, if entire string contains alphabets
-isdigit(): it returns true, if entire string contains digits
-isalnum(): it returns true, if entire string contains either alphabets or digits
-startswith(substring): it returns true if it starts with substring
-endswith(substring): it returns true if it ends with substring
```

```python
print("CODEgnan".isupper())
print("CODEgnan".islower())
print("CODEGNAN".isupper())
print("CODEGNAN1234".isalnum())
print("832822".isdigit())
print("832822".isalnum())
print("code".startswith("c"))
print("code".startswith("C"))
```
