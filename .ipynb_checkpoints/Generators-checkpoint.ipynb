Generators:
-is a special type of function.
-it allows you to generate one value at a time
-it takes less memory because it doesn't care about previous value space
-also called lazy function beacuse it evaluates slowly one after another value
-generators are more memory efficient.

>The generator function returns a iterable object.
>syntax:
    def generator_name(parameters):
        #block code
        yield val/exp
    res = generator_name(arg)
    print(next(res))            #generators obj with next function then only it executes from starting to reach yield statement
    print(next(res))            #it starts executing from previous pause

#if you call generator object again with next() it starts executing from previous wherever it left off.

```python
def gen(n):
    for i in range(1,n+1):
        yield i

n = int(input())       
res = generator(n)   #it doesn't start executes generator
print(next(res))     #here generator function execution starts it executes until it reaches to yield statement.
print(next(res))
print(next(res))
print(next(res))
print(next(res))
print(next(res))
print(next(res))




```

```python
#return 1 to n numbers using function

def numbers(n):
    res = []
    for i in range(1,n+1):
        res.append(i)
    return res
print(numbers(n))
```

```python
def gen(n):
    for i in range(1,n+1):
        yield i

n = int(input())       
res = gen(n)   
print(res)
for value in res:
    print(value)


```

```python
import time
import sys
s = time.time()          #start time 
def gen(n):
    for i in range(1,n+1):
        yield i**3

n = int(input())       
res = gen(n)   
print(res)
for value in res:
    print(value)
e = time.time()      #end time
print(e-s)                 #end time - start time
```

```python
#return 1 to n numbers using function
import sys
import time
s = time.time()  
def numbers(n):
    res = []
    for i in range(1,n+1):
        res.append(i**3)
    return res

n = int(input())
print(numbers(n))


e = time.time()
print(e-s)
```
