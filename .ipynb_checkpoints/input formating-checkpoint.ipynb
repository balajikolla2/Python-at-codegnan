Input Formating:

1.Integer reading from user by default input() is in string format.

```python
user = input("Enter you fav number:")
integer = int(user)
print(integer)
print(type(integer))
```

```python
#Reading float value from user
value = float(input("Hey! Tell me a one fav number:"))
print(value)
print(type(value))
```

```python
#complex reading from user
value = complex(int(input("Enter you 1st number: "),int(input("Enter your 2nd number: "))))
print(value)
print(type(value))
```

```python
#boolean reading to integer
value = bool(int(input("Hey John! Please enter number: ")))
print(value)
```

map(): This function applies functionality on each element in iterable

map syntax: map(function_name,iterable)

iterable means lists tuples sets

```python
list1= list(map(int, input().split()))
print(list1)
```
