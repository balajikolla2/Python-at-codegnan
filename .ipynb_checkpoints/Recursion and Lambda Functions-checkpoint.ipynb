Recursion:
    -A function calls itself either directly or indirectly.

Types of Recursion:
    -1.Direct recursion
    -2.Indirect recursion

1.Direct Recursion: A function calls itself until it reaches to base condition

Main blocks: >Base condition
             >Recursive call

1.Base condition:
    -The Base condition is used to stop the infinite recursive call.

2.Recursive Call:
    -The function call recursively until it reaches to base condition

```python
def fun(x):
    if x == 0:
        return 0 
    x = x - 1 
    fun(x)
    print(x)

fun(11)
```

```python
#print 1 to n using recursion
def fun(x):
    if x == 0:
        return 0 
    fun(x-1)
    print(x)
fun(4)
```

```python
#print 1 to n using recursion
def fun(x):
    if x == 0:
        return 0 
    print(x)
    fun(x-1)
fun(4)
```

```python
#factorial of a number

def fact(n):
    if n == 0 or n == 1:
        return 1
    return n*fact(n-1)

    
n = int(input())
fact(n)
```

```python
#to print the len

def len_find(n):
    length = len(n)
    return length


n = input()
len_find(n)
```

2. Indirect Recursion: Function 'a' calls function 'b' and function 'b' calls function 'a'

Lambda function or anoymous functions:

    -The function doesn't have function name,those functions called as anonymous function.
    -The anonymous function can implement with 'lambda' keyword are also called 'single line function'

    Syntax:
        lambda arguments: Expression

```python
#by using 'lambda fn' to fing square of a number

s = lambda x: x ** 2
print(s(5))
```

```python
#find the square of list of integers using map and lambda

arr = [1,2,3,4,5]
print(list(map(lambda x: x**2 , arr)))
```

```python
#using filter and print even numbers
arr = [1,2,3,4,5]
print(list(filter(lambda x: x%2==0 , arr)))
```

```python
#single line if-else:

syntax: 
    True_exp if condition else False_exp  
```

```python
#single line even or odd 'if-else;

n = 10
res = 'even' if n%2==0 else 'odd'
print(res)
```

```python
arr = [1,2,3,4,5]
print(list(map(lambda x: x%2==0 , arr)))
```

```python
arr = [1,2,3,4,5]
#1. -odd
#2. -even
arr = [1,2,3,4,5]
print(list(map(lambda x:f"{x} - even" if x%2==0 else f"{x} -odd", arr)))
```

```python
#to find the frequency of each_char

string = input()
count = {}
for char in string:
    if char in count:
        count[char] += 1 
    else:
        count[char] = 1 

print(count)
```

```python
#to find the frequency of each_char

string = input()
count = {}
for char in string:
    count[char] = count.get(char,0) + 1

print(count)
```
