Scope: A Scope represents region of a program and where variable or name can access

Types of scopes:
1.Local Scope
2.Enclosing Scope
3.Global Scope
4.Built in Scope

1.Local Scope: Define a variable inside the function.
example:

    def fun():
        x = 10         #defined inside a function
        print(x)       #local variable

    fun()

2.Enclosing Scope: A variable is defined in outer function and accessing in inner function(nested function)



def outer():
    y = 20        #enclosing variable
    def inner():
        z = 10    #local variable
        print(z,y)
    inner()
outer() #output 10,20


3.Global Scope:


x = 10            #Global Variable
def outer():
    y = 20        #enclosing variable
    def inner():
        z = 30    #local variable
        print(x,y,z)
    inner()
outer() #output 10,20,30



4,Built in scope:
    The python provide the built in functions like print(),len(),sum(),min(),max() ....etc

Note: When ever if you try to access any variable or name it starts checks in 'LEGB' Rule

L - Local Scope
E - Enclosing Scope
G - Global Scope
B - Built in scope

```python
#scope of variables
#1.Local scope
#2.Enclosing scope
#3.Global scope
#4.Builtin scope
x = 10    #global scope variable
def outer():
    y =20   #enclosing scope variable
    def inner():
        z = 30      #local scope varible
        print(z,y,x)     #builtin scope function
    inner()
outer()
```

```python
#using global scope variable in local scope
x = 100
def outer():
    y = x + 1 
    print("Local scope variable value:",y)
outer()
print("Global scope variable value:",x)
```

```python
#updating the global scope variable value in local scope
#can't update global scope variable in local scope directly

x = 100
def outer():
    x = x + 1 
    print("Local scope variable value:",y)
outer()
print("Global scope variable value:",x)
```

```python
#updating the global scope variable value in local scope by using 'global' keyword

x = 100
def outer():
    global x
    x = x + 1
    print("local scope variable value:",x)
outer()
print("global scope variable value:",x)
```

```python
#updating the enclosing scope variable value in local scope 
x = 10    #global scope variable
def outer():
    y =20   #enclosing scope variable
    def inner():
        z = 30
        y = y + z
        print(z,y,x)     #builtin scope function
    inner()
outer()
```

```python
#updating the enclosing scope variable value in local scope by using nonlocal
x = 10    #global scope variable
def outer():
    y =20   #enclosing scope variable
    def inner():
        nonlocal y
        z = 30
        y = y + z
        print(z,y,x)     #builtin scope function
    inner()
    print("Enclosing scope variable value:",y)
outer()
```
