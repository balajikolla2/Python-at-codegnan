```python
#to print hallow square

n = int(input())                                         #i==0 for top border
for i in range(n):                                       #i==4 means i==n-1 for bottom border
    for j in range(n):                                   #j==0 for left border
        if (i==0 or i==n-1 or j==0 or j==n-1):           #j==n-1 for right border
            print("*",end = " " )
        else:
            print(" ",end = " ")
    print()
```

```python
#number increment
n = int(input())
num = 1
for i in range(1,n+1):
    for j in range(1,n+1):   #in this case if i write n+1 instead of i+1 then in each line it prints 5 values in each row
        print(num,end = " ")
        num += 1
    print()
```

```python
n = int(input())
num = 1
for i in range(1,n+1):
    for j in range(1,i+1):   #in this case if i write n+1 instead of i+1 then in each line it prints 5 values in each row
        print(num,end = " ")
        num += 1
    print()
```

```python
n = int(input())
num = n*(n+1)//2          #num for calculate the value it gives 15 for 5(input) example
for i in range(1,n+1):
    for j in range(i):   
        print(num,end = " ")
        num -= 1
    print()
```

```python
n = 5
for i in range(n):
    for j in range(n):
        if i == 0 or i == n-1 or j == n//2:
            print("*", end="")
        else:
            print(" ", end="")
    print()
```
