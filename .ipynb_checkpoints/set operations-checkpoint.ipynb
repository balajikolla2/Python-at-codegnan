## SET OPERATIONS
UNION: IT RETURN ALL ELEMENTS FROM ALL SETS WITHOUT DUPLICATES
INTERSECTION: IT RETURNS ONLY COMMON ELEMENTS FROM THE SET
A-B : IT RETURNS ONLY ELEMENTS IN SET A
SYMMETRIC DIFFERENCE: IT RETURNS NON COMMON ELEMENTS 


SET METHODS

```python
#Element adding method
#add(ele): It adds the element to set if element not in set,otherwise nothing
s = {1,2,3,4,5,5,4}
print(s)  #it removes duplicate values
s.add(100)
s.add(5)
s.add("codegnan")  #add strings in sets
print(s)
```

```python
#Element removing methods
#remove(item): It removes the item if it exists in set , otherwise 'KeyError'
#discard(item): It removes the item if it exists in set , otherwise nothing
s = {1,2,3,4,5,5,4}
s.discard(5)
print(s)
```

```python
#empty set
s = { }
s1 = set()
print(type(s))
print(type(s1))
```

```python
##SET OPERATIONS
# union(set): It returns the all unque elements from all sets
s1 = {1,3,5,7}
s2 = {1,2,4,5,7}
union = s1.union(s2)
print(union)
```

```python
s1 = {1,3,5,7}
s2 = {1,2,4,5,7}
s3 = {7,8,9}
union = s1.union(s2,s3)
print(union)
res = s1 | s2 | s3
print(res)
```

```python
#intersection
s1 = {1,3,5,7}
s2 = {1,2,4,5,7}
s3 = {7,8,9}
i = s1.intersection(s2,s3) #it returns only common elements
print(i)
res = s1 & s2 & s3
print(res)
```

```python
s1 = {1,3,5,}
s3 = {7,8,9}
i = s1.intersection(s3)  #if there is no common element then it returns the empty set
print(i)
```

```python
# difference(set): It returns the set elements those are not in another set
s1 = {1,3,5,7}
s2 = {1,2,4,5,7}
s3 = {7,8,9}
i = s1.difference(s2,s3) #it returns only common elements
print(i)
res = s1 - s2 - s3
print(res)
print((s1 | s2) - s3)
```

```python
# symmetric_difference(set): It returns the non common elements from all sets
s1 = {1,3,5,7}
s2 = {1,2,4,5,7}
s3 = {7,8,9}
s_diff = s1.symmetric_difference(s2) #it returns only common elements
print(s_diff)
res = s1 ^ s2 ^ s3
print(res)
```

```python
#issubset(set)
#issuperset(set)
#isdisjoint(set)

##these returns only boolean values(T/F)
s1 = {1,3,5,7}
s2 = {1,3,5,7}
sub = s1.issubset(s2)
supers = s1.issuperset(s2)
disj = s1.isdisjoint(s2)
print(sub)
print(supers)
print(disj)
```

```python
s1 = {1,3,5,7}
s2 = {1,2,4,5,7}
s3 = {7,8,9}
s1.intersection_update(s2,s3)
print(s1,s2,s3)
```

```python
s1 = {1,3,5,7}
s2 = {1,2,4,5,7}
s3 = {7,8,9}
s1.difference_update(s2,s3)
print(s1,s2,s3)
```
