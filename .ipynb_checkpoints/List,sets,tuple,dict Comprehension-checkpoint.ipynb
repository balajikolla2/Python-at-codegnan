List Comprehension:

-This is the one of way to create a list with an expression

Syntax:

[expression for iterator in iterable if condition]

```python
#Even numbers
n = int(input())
res = [x for x in range(0,n+1) if x%2==0 ]
print(res)
```

```python
#generate 1-n numbers using list comprehension

n = int(input())
res = [x for x in range(1,n+1)]
print(res)
```

```python
#find cubic numbers

n = int(input())
res = [x**3 for x in range(1,n+1)]
print(res)
```

```python
#find the odd numbers between n1 and n2

n1,n2 = map(int,input().split())
res = [x for x in range(n1,n2+1) if x%2==1 ]
print(res)
```

```python
# in given arr of integers perform following operations
#if even - find square of number
#if odd - find squareroot value(upto 2 value)


arr = list(map(int,input().split()))
res = [x**2 if x%2==0 else round(x**0.5,2) for x in arr]  #if 'if else' is exist then we write before for loop 
print(res)
```

```python
#2 dices combination values
res = [ (x,y) for x in range(1,7) for y in range(1,7)]
print(res)
```

```python
#2 dices combination values if sum of values div by 3 
res = [ (x,y) for x in range(1,7) for y in range(1,7) if (x+y)%3==0 ]    #1st loop is outer loop next loop is inner loop
print(res)
```

```python
#count the no of vowels exists in string
string = input()
res = [char for char in string.lower() if char in "aeiou"  ]
print(res)
print(len(res))
```

```python
#generate 1-n numbers using set comprehension

n = int(input())
res = {x for x in range(1,n+1)}
print(res)
print(type(res))
```

```python
#dictionary comprehension

n = int(input())
res = {x:x**2 for x in range(1,n+1)}
print(res)
print(type(res))
```

```python
#tuple comprehension

n = int(input())
res = tuple(x for x in range(1,n+1)) #here we pass tuple as function if not it generators object
print(res)
print(type(res))
```

```python
s1 = input()
rev = ""
for each_char in s1:
    rev = each_char + rev 
print(rev)
```
