##SETS
    -is a unordered collection of different datatype unique elements.
    -are mutable
    -defines with '{ }'
Characteristics:
    -is unordered collection
    -elements are unique
    -does not support indexing
    -are mutable
    -allows hashable datatypes as elements (lists is unhashable so not can be used in sets)
    -can be used as dictionary keys

Set Characteristics

```python
#set unordered collection 
#sets return unique elements does not allow duplicates
s = {1,2,3,4,2,3,1,'hi','b',2.56,(10,20,30)}
print(s)
print(s[0]) #set does not support indexing
```

```python
s = {1,2,3,4,2,3,1,'hi','b',2.56,(10,20,30),[10,20,30],34}
print(s) #because list is unhashable type
```
