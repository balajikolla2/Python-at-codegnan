LOOP CONTROL STATEMENTS 
1.continue
2.break

1.continue: after executing of 'continue' keyword execution immediately it goes to next iteration and skips the current iteration after continue keyword statements.

```python
#Loop control statements
#continue
for i in range(1,12):
    if i%2==0:
        continue
    print(i)
```

2.break:
    -after executing of 'break' keyword immediately execution goes to out of loop.
    -that means loop will terminate
    -the loops else block executes when break keyword not executed

```python
#finding the index value for a given num in list

list2 = [12,32,45,32,65,76,90,66]
target = 65
n = len(list2)
for i in range(n):
    if list2[i] == target:
        print(i)
        break
else:
    print("-1")
```

```python
#here input is reading from the user and find the index for the given number
list2 = list(map(int, input().split()))
target = 65
n = len(list2)
for i in range(n):
    if list2[i] == target:
        
        print(i)
        break
else:
    print("-1")
```

```python
#factors of prime numbers:
n = int(input())
factors = list()
for i in range(1,n+1):
    if (n%i==0):
        factors.append(i)
print(factors)
```
