```python
n = int(input())
if (n%2==0) and (n>=0):
    print(" Positive Even")
elif (n%2==1) and (n>=0):
    print("Positive odd")
elif (n%2==0) and (n<=0):
    print("neg even")
elif (n%2==1) and (n<=0):
    print("neg odd")
```

```python
#pos even odd,neg even odd using nested 
n = int(input())
if n>0:
    if (n%2==0):
        print("pos even")
    else:
        print("pos odd")
else:
    if (n%2==0):
        print("neg even")
    else:
        print("neg odd")
```
