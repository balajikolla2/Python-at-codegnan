TYPE CONVERTION:
    -Converting one data type format into another datatype format this is called as type convertion.
Two Types:
    -implicit type convertion
    -explicit type convertion(type casting)
Implicit type conversion:
    -The python interpreter automatically converts the datatype because it prevents the dataloss.
    -implicit type conversion can happens in arthematic operations
Explicit type conversion:(Type Casting)
    -Here type conversion happens by using python built in datatype functions.

Builtin Data type Functions:
1.integers = int()                              a = 5.6
2.float    = float()                            b = int(5.6)
3.complex  = complex()                          print(type(b)) 
4.boolean  = bool()
5.string   = str()
6.list     = list()
7.tuple    = tuple()
8.set      = set()
9.dictionary=dict()

In explicit: primitive to primitive and non primitive to non primitive

Python: is a case sensitive means Capital and small are not same

```python
#integer to float,boolean,string
number = 10
print(float(number))
print(bool(number))
print(str(number))
print(complex(number))
```

```python
number = 10,45,65
print(list(number))   #converting to list
print(set(number))    #converting to set()
```

```python
number = 10,45,65
print(list(number))
print(type(number))
```
