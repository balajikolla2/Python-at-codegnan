```python
#list to string,tuple,set
fruits = ['apple','banana','orange']
string = " ".join(fruits)  #join: combines all items in list into string
print(string)
```

```python
num = [10,20,40,50,60]
output = tuple(num)
print(output)
```

```python
num = [10,20,40,50,60]
output = set(num)
print(output)
```

```python
#tuple to string,list,set
names = ("balaji","laasya","hemanth")
string = " ".join(names)
print(string)
```

```python
num = [10,20,40,50,60]
output = list(num)
print(output)
```

```python
num = [10,20,40,50,60]
output = set(num)
print(output)
```

```python
#dictionary to string,list,set,tuple
student = {'name':'balaji','age':23,'course':'python'}  #converting only keys to string
string = " ".join(student)
print(string)
```

```python
student = {'name':'balaji','age':23,'course':'python'}  #converting only keys to list
string = list(student)
print(string)
```

```python
student = {'name':'balaji','age':23,'course':'python'}  #dic keys become only sets
string = set(student)
print(string)
```

```python
student = {'name':'balaji','age':23,'course':'python'}  #dic keys become only tuples
string = tuple(student)
print(string)
```
