##   DICTIONARY:
    -is a ordered collection key - value(i.e items) pairs.
    -it is mutable
    -defines with '{ }' i.e, 
    -format- {key : value}
    CHARACTERISTICS:
    -is a ordered collection
    -is mutable
    -does not support indexing
    -dictionary keys are unique
    -values allows duplicates
    -list can't be used as dictionary key because unhashable.

Hashing: it is a storing mechanism

```python
# Dictionary ordered collection
details =  {'name': 'balaji','course': 'python','batch': 57,'age': 23}
print(details)
```

```python
details =  {'name': 'balaji','course': 'python','batch': 57}
details['batch']
```

```python
#Dictionary values access by using keys
#updating dictionary values
#adding new items to dicionary
details =  {'name': 'balaji','course': 'python','batch': 57,'age': 23}
print(details['course'])
#updating the dictionary values
details['age'] = 25
print(details['age'])
#adding new items
details['location'] = 'hyderabad'
print(details)
details['age'] = details['age'] + 4 
print(details)
```

```python
# dictionary keys are unique
details =  {'name': 'balaji','age' : 34,'course': 'python','batch': 57,'age': 23}
print(details)
```

```python
# List can't be used as dictionary
details =  {'name': 'balaji','age' : 34,'course': 'python',[1,2,3] : 57,'age': 23,}
print(details)
```

###Dictionary Methods

- get(key,default_value): It returns the value if key is exists in dictionary,otherwise it returns default value

```python
details =  {'name': 'balaji', 'course': 'python', 'batch': 57, 'age': 25, 'location': 'hyderabad'}
print(details)
print(details.get('name'))
print(details.get('marks',"marks not available")) 
print(details.get('marks'))                       #this one returns none by default if key is not exist in dictionary
```

- update(new_dictionary) : This method updates the values if new_dictionary keys already exists in dictionary,otherwise it adds new items to dictionary

```python
details =  {'name': 'balaji', 'course': 'python', 'batch': 57, 'age': 25}
details.update({'batch': 58,'course': 'Java', 'location': 'hyderabad'})
print(details)
```

- keys()   : This method returns the list dictionary keys
- values() : This method returns the list dictionary values
- items()  : This method returns the list of tuple of key-value pairs

```python
details =  {'name': 'balaji', 'course': 'python', 'batch': 57, 'age': 25}
keys = details.keys()
print(type(keys))
print(keys)
print(details.values())
print(details.items())
```

- popitem(): It return and removes the last item as tuple
- pop(key) : This method removes the specified key item and return the specified keys value,otherwise KeyError

```python
details =  {'name': 'balaji', 'course': 'python', 'batch': 57, 'age': 25}
item = details.popitem()
print(item)
print(details)
val = details.pop('batch')
print(val)
print(details)
```

```python
#membership checking
details =  {'name': 'balaji', 'course': 'python', 'batch': 57, 'age': 25}
print('babu' in details)
print('babu' not in details)
print(len(details)) //To find length
```
